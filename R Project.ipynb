```python
!pip install rdkit

# Imports
#------------------------------------------------------------------------------------------------------#
import pandas as pd 
#------------------------------------------------------------------------------------------------------#
from rdkit import Chem, RDLogger
from rdkit.Chem import Descriptors, MolSurf, rdMolDescriptors, AllChem, Crippen, PandasTools, Lipinski
#------------------------------------------------------------------------------------------------------#
from google.colab import files  # Poder importar arquivos
from google.colab import drive  # Conectar colab com o drive
drive.mount('/content/drive')
```

__________________________________________________________


```python
chebi = '/content/drive/MyDrive/ChEBI_complete.sdf' # Lendo a base de dados upada do google drive 
```

```python
all_mols = Chem.SDMolSupplier(chebi)  # Vai puxar os mols da database ChEbi
```

```python
# Criou-se uma lista para cada coluna em que se precisava fazer cálculo  
all_mols = Chem.SDMolSupplier(chebi)  # Extração dos mols de chebi
SMILES = []
NumAtoms=[]
ExactMolWt=[]
Mol_LogP = []
NumHAcceptors = []
NumHDonors = []
Ring_Count = []
TPSA =[]
NumRotableBonds = []
Molecular_formula = []
for x in range (0, len(all_mols)):  # O x vai percorrer a condição(if) e executar ela len(all_mols)
  if all_mols[x] is None: continue  # Se o x dentro de all_mols for None ele vai continuar e executar os cálculos
  NumAtoms.append(all_mols[x].GetNumAtoms())
  SMILES.append(Chem.MolToSmiles(all_mols[x]))
  ExactMolWt.append(Descriptors.ExactMolWt(all_mols[x]))
  TPSA.append(Descriptors.TPSA(all_mols[x]))
  Mol_LogP.append(Crippen.MolLogP(all_mols[x]))
  NumHAcceptors.append(Lipinski.NumHAcceptors(all_mols[x]))
  NumHDonors.append(Lipinski.NumHDonors(all_mols[x]))
  Ring_Count.append(Lipinski.RingCount(all_mols[x]))
  NumRotableBonds.append(Lipinski.NumRotatableBonds(all_mols[x]))
  Molecular_formula.append(rdMolDescriptors.CalcMolFormula(all_mols[x]))
```

```python
dataSDF = '/content/ChEBI_complete.sdf'  # Leitura da base de dados por upload direto na máquina
```

```python
ChEbi = PandasTools.LoadSDF(dataSDF)  # Lê o sdf e transforma a variavel em data frame 
```

```python
Chebi_ID = ChEbi['ChEBI ID']
```

__________________________________________________________


```python
# Adicionando as listas em um dicionário
i = {'ChEbi_ID': Chebi_ID,
     'SMILES': SMILES,
    'MolecularFormula': Molecular_formula,
    'NumAtoms': NumAtoms,
    'ExactMolWt': ExactMolWt,
    'NumRotableBonds': NumRotableBonds,
    'MolLogp': Mol_LogP,
    'RingCount': Ring_Count,
    'NumHAcceptors': NumHAcceptors,
    'TPSA': TPSA,
    'NumHDonors': NumHDonors,

}
# Criando um data frame a partir do dict(i) e transformando ele em csv
pd.DataFrame(i).to_csv('base de dados.csv', sep = ';')
```

```python
MoleculeDescription = pd.DataFrame(data=i)  # Tornando o dict em um data frame que pode ser consultado como tabela
```
